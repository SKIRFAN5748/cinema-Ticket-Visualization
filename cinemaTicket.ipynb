```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset=pd.read_csv('cinemaTicket_Ref.csv')
dataset.head()
```

```python
plt.figure(figsize=(20,10))
plt.xlabel("tickets_sold")
plt.ylabel("month")
sns.scatterplot(x='month',y='tickets_sold',hue='day',data=dataset)
```

```python
plt.figure(figsize=(20,10))
plt.xlabel("cinema_code")
plt.ylabel("total_sales")
sns.scatterplot(x='cinema_code',y='total_sales',hue='day',data=dataset)
```

```python
plt.figure(figsize=(20,10))
plt.xlabel("cinema_code")
plt.ylabel("ticket_price")
sns.scatterplot(x='cinema_code',y='ticket_price',hue='ticket_price',data=dataset)
```

```python
pd3=dataset['capacity']
sns.countplot(pd3)
```

```python
pivot1=\
dataset.pivot_table(index='day',values='tickets_sold',aggfunc=np.sum)
plt.figure(figsize=(20,15))
pivot1.plot(kind='bar',color='yellow')
plt.xlabel('day')
plt.ylabel('tickets_sold')
plt.show()
```

```python
pivot2=\
dataset.pivot_table(index='month',values='tickets_sold',aggfunc=np.sum)
plt.figure(figsize=(20,15))
pivot1.plot(kind='bar',color='red')
plt.xlabel('month')
plt.ylabel('tickets_sold')
plt.show()
```

```python
pivot3=\
dataset.pivot_table(index='cinema_code',values='ticket_price',aggfunc=np.mean)
plt.figure(figsize=(15,20))
pivot3.plot(kind='bar',color='green')
plt.xlabel('cinema_code')
plt.ylabel('ticket_price')
plt.show()
```

```python
pivot4=\
dataset.pivot_table(index='cinema_code',values='ticket_use',aggfunc=np.mean)
plt.figure(figsize=(15,20))
pivot3.plot(kind='bar',color='green')
plt.xlabel('cinema_code')
plt.ylabel('ticket_use')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.xlabel("cinema_code")
plt.ylabel("capacity")
sns.scatterplot(x='cinema_code',y='capacity',hue='day',data=dataset)
```
